```python
import pandas as pd

# Read the first TSV file
df1 = pd.read_csv('valid_preprocess.tsv', sep='\t')

# Read the second TSV file
df2 = pd.read_csv('train_preprocess.tsv', sep='\t')

```

```python
# adding column name to the respective columns
df1.columns =['text', 'label']
df2.columns =['text', 'label']
```

```python
df1
```

```python

# Perform inner join based on a common identifier column
merged_df = pd.concat([df1, df2], ignore_index=True)

# Save the merged dataframe to a new TSV file
merged_df.to_csv('merged_train_valid_file.tsv', sep='\t', index=False)
```
